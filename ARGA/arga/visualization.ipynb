```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_excel(r'results/clustering.xlsx')
df.set_index(df['epoch'], inplace=True)
print(df)
```

```python
df_ = df.loc[:, ['ARGA-cora-ACC','ARGA-cora-f1_macro', 'ARGA-cora-NMI',
                'ARVGA-cora-ACC','ARVGA-cora-f1_macro', 'ARVGA-cora-NMI']]

df_.plot()
plt.title("clustering - cora")
plt.xlabel("epochs")
plt.ylabel("val")
plt.show()
```

```python
df_ = df.loc[:, ['ARGA-citeseer-ACC','ARGA-citeseer-f1_macro', 'ARGA-citeseer-NMI',
                 'ARVGA-citeseer-ACC','ARVGA-citeseer-f1_macro', 'ARVGA-citeseer-NMI']]

df_.plot()
plt.title("clustering - citeseer")
plt.xlabel("epochs")
plt.ylabel("val")
plt.show()
```

```python
df = pd.read_excel(r'results/link_prediction.xlsx')
df.set_index(df['epoch'], inplace=True)
print(df)
```

```python
df_ = df.loc[:100, ['ARGA-cora-train_loss','ARGA-cora-val_roc', 'ARGA-cora-val_ap',
                    'ARVGA-cora-train_loss','ARVGA-cora-val_roc', 'ARVGA-cora-val_ap']]

df_.plot()
plt.title("link_prediction - cora")
plt.xlabel("epochs")
plt.ylabel("val")

axes = plt.gca()
axes.set_ylim([0.2,1.3])

plt.legend(loc='lower right')

plt.show()
```

```python
df_ = df.loc[:100, ['ARGA-citeseer-train_loss','ARGA-citeseer-val_roc', 'ARGA-citeseer-val_ap',
                    'ARVGA-citeseer-train_loss','ARVGA-citeseer-val_roc', 'ARVGA-citeseer-val_ap']]

df_.plot()
plt.title("link_prediction - citeseer")
plt.xlabel("epochs")
plt.ylabel("val")

axes = plt.gca()
axes.set_ylim([0.2,1.3])

plt.legend(loc='lower right')

plt.show()
```

```python
import numpy as np
emb = np.load('results/cora-arga-emb.npy')
label = np.load('results/cora-arga-labels.npy')
```

```python
from sklearn.manifold import TSNE

model = TSNE(learning_rate=100, init='pca')
transformed = model.fit_transform(emb)

xs = transformed[:,0]
ys = transformed[:,1]
plt.scatter(xs,ys,c=label)

plt.title("tSNE - ARGA")
plt.xlabel("x")
plt.ylabel("y")

plt.rcParams["figure.figsize"] = (6,6)

plt.show()
```

```python
xs = emb[:,0]
ys = emb[:,1]

axes = plt.gca()
axes.set_xlim([-1.2,1.2])
axes.set_ylim([-1.2,1.2])

plt.title("latent space - ARGA")
plt.xlabel("$z_0$")
plt.ylabel("$z_1$")

plt.rcParams["figure.figsize"] = (6,6)

plt.scatter(xs,ys,c=label)
```
